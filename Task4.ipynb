```python
1 #mobile recharge validation system
def recharge_validation():
  minimum_amount=50
  plans=[199,299,399,599]
  while True:
    try:
      amount=int(input("Enter the amount:"))
      if amount<50:
        print("Amount should be greater than 50")
      elif amount not in plans:
        print(f"{amount} not in plans")
      else:
        print(f"{amount} is valid plan")
    except ValueError:
        print("Invalid input.Please enter a valid amount.")
        #retry logic
        retry=input("Do you want to retry? (yes/no):")
        if retry.lower()!="yes":
          print("transaction cancelled")
        break

recharge_validation()

```

```python
3 #student result processing system
def student_result():
  marks=[80,55,74,90,99,35,64]
  average=sum(marks)/len(marks)
  print(f"Average marks:{average}")
  if average>=50:
    print("pass")
  else:
    print("fail")
  # Iterate through each mark in the 'marks' list
  result=["pass" if mark>=50 else "fail" for mark in marks ]
  print(result)
student_result()
```

```python
#cab fare estimator with retry option

def calculate_cab_fare(distance_km, is_peak_hour):
  base_fare = 50
  cost_per_km = 12
  peak_hour_surcharge_percent = 0.25

  fare = base_fare + (distance_km * cost_per_km)

  if is_peak_hour:
    fare *= (1 + peak_hour_surcharge_percent)

  return fare

while True:
  try:
    distance_input = input("Enter distance in kilometers (or 'exit' to quit): ")
    if distance_input.lower() == 'exit':
      print("Exiting cab fare calculator. Goodbye!")
      break

    distance_km = float(distance_input)
    if distance_km < 0:
      print("Distance cannot be negative. Please try again.")
      continue

    peak_hour_input = input("Is it a peak hour? (yes/no): ")
    is_peak_hour = peak_hour_input.lower() == 'yes'

    final_fare = calculate_cab_fare(distance_km, is_peak_hour)
    print(f"The estimated cab fare is: â‚¹{final_fare:.2f}")

  except ValueError:
    print("Invalid input. Please enter a numeric value for distance.")
  except Exception as e:
    print(f"An unexpected error occurred: {e}")

  retry = input("Do you want to calculate another fare? (yes/no): ")
  if retry.lower() != 'yes':
    print("Thank you for using the cab fare calculator!")
    break
```

```python
5 #employee attendance eligibility checker
def employee_attendance():
  attendance_list=["present","absent","present","absent","present"]
  present_days=attendance_list.count("present")
  total_days=len(attendance_list)
  attendance_percentage = (present_days / total_days) * 100
  print(f"attendance :{attendance_percentage  }")
  if attendance_percentage >=75:
    print("Eligible ")
  else:
    print("Not eligible ")
    return attendance_percentage
employee_attendance()



```

```python
6 #password strength checker
def strength_checker():
  password=input("Enter the password:")
  if len(password)<8:
    print("Password is too short , minimum length is 8 ")
  elif  not any(char.isdigit() for char in password):
    print("Password should contain at least one digit")
  elif not any("(@#$) for char in password"):
    print("Password should contain at least one specia case charachter")
  else:
    print("Password is strong")
strength_checker()
```
