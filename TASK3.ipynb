```python
import collections

#search query keyword analysis
query="Buy mobile phone buy phone online"
# Convert the query to lowercase and split it into words
words = query.lower().split()

# Use collections.Counter to count the frequency of each word
word_counts = collections.Counter(words)

# Print the word counts
print("Word frequencies:", word_counts)

# Identify keywords searched more than once
keywords_searched_more_than_once = [word for word, count in word_counts.items() if count > 1]
print("Keywords searched more than once:", keywords_searched_more_than_once)
```

```python
3 #sensor data validation
sensor_readings=[3,4,7,8,10,12,5]
even_readings = [] # Initialize even_readings as an empty list
for reading in sensor_readings:
    if reading%2==0:
        even_readings.append(reading)
print("Even readings:", even_readings) # Print the final list outside the loop
```

```python
# Store sensor readings as (hour, value) for even readings
sensor_data_even_tuples = set()
for i, reading_value in enumerate(sensor_readings):
    if reading_value % 2 == 0: # Check if the value is even
        sensor_data_even_tuples.add((i, reading_value))

print("Sensor data with even readings as (hour, value) tuples:", sensor_data_even_tuples)
```

```python
4 #email domain usage analysis
emails=["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
domain_usage={}
for email in emails:
    domain=email.split("@")[1]
    if domain in domain_usage:
        domain_usage[domain]+=1
    else:
        domain_usage[domain]=1

# Calculate total number of emails
total_emails = len(emails)

# Calculate and print domain usage as percentages
print("Email Domain Usage:")
for domain, count in domain_usage.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}:{percentage:.0f}%")
```

```python
import collections

6 #duplicate userid detection
user_ids=["user1","user2","user1","user3","user1","user3"]
print("Original user IDs:", user_ids)

# Count the frequency of each user ID
user_id_counts = collections.Counter(user_ids)

print("Duplicate user IDs and their counts:")
# Identify and print duplicate user IDs along with their counts
for user_id, count in user_id_counts.items():
    if count > 1:
        print(f"{user_id}: {count} times")
```
