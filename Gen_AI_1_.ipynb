```python
from langchain.prompts import ChatPromptTemplate

def get_chat_prompt(role, topic):
    system_roles = {
            "teacher": "You are a helpful teacher who explains concepts clearly.",
                    "interviewer": "You are an interviewer who asks insightful questions.",
                            "motivator": "You are a motivational speaker who inspires learners."
                                }

                                    chat_template = ChatPromptTemplate.from_messages([
                                            ("system", system_roles.get(role, "You are a helpful assistant.")),
                                                    ("user", "Explain {topic}")
                                                        ])

                                                            return chat_template.format_messages(topic=topic)

                                                            # Example
                                                            messages = get_chat_prompt("teacher", "Neural Networks")
                                                            print(messages)
```

```python
def validate_inputs(audience, tone):
      valid_audience = ["beginner", "intermediate", "expert"]
          valid_tone = ["formal", "casual", "fun"]

              if audience not in valid_audience:
                      print("Invalid audience! Defaulting to 'beginner'")
                              audience = "beginner"

                                  if tone not in valid_tone:
                                          print("Invalid tone! Defaulting to 'formal'")
                                                  tone = "formal"

                                                      return audience, tone
```

```python
def generate_prompt(topic, audience, tone, style):
      audience, tone = validate_inputs(audience, tone)

          base_template = PromptTemplate(
                  input_variables=["topic", "audience", "tone", "style"],
                          template="Explain {topic} for {audience} in a {tone} {style} style"
                              )

                                  style_map = {
                                          "teaching": "teaching",
                                                  "interview": "interview by",
                                                          "storytelling": "storytelling"
                                                              }

                                                                  style_value = style_map.get(style, "teaching")

                                                                      return base_template.format(
                                                                              topic=topic,
                                                                                      audience=audience,
                                                                                              tone=tone,
                                                                                                      style=style_value
                                                                                                          )

                                                                                                          # Example
                                                                                                          print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
