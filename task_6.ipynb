```python
1#smart parking lot management system
def alot_parking(capacity,logs):
  park_count=0
  for action in logs:
    if action=='IN':
      if park_count < capacity: # Check for capacity before allowing 'IN'
        park_count+=1
      else:
        print("Alert:FULL! cannot allot vehicle (lot is full)")
    elif action=="OUT": # Corrected indentation for elif
      if park_count>0:
        park_count-=1
      else:
        print("Alert: No vehicles to exit (lot is empty)") # Added for clarity
    else: # Handle unrecognized actions
      print(f"Alert: Unrecognized action '{action}'")

  if park_count >= capacity: # Use >= for safety
    status="FULL"
  else:
    status="Available"
  return park_count, status # Return the calculated count and status

parking_capacity=50
vehicle_logs=["IN","IN","IN","OUT","IN","IN","OUT"]

# Assign the returned values to count and current_status
count, current_status = alot_parking(parking_capacity,vehicle_logs)

print(f"currently parked vehicles:{count}")
print(f"parking status:{current_status}")
print("this is smart parking allotment system")
```

```python
2#online food delivery time estimator
def delivery_time(distance,weather,trafficlevel):
  prep_time=15
  travel_time=distance*5
  if trafficlevel.lower()=="high":
    travel_time*=1.25
  elif trafficlevel.lower()=="medium":
    travel_time*=1.10
  if weather.lower()=="rainy":
    travel_time*=1.10
  elif weather.lower()=="stormy":
    travel_time*=1.5
  total=prep_time+travel_time
  result= round(total)
  return result
#sample input
distance=8
weather="rainy" # Changed to 'rainy'
trafficlevel="high" # Changed to 'high'
eta= delivery_time(distance,weather,trafficlevel)
print(f"Distance: {distance}km")
print(f"Weather: {weather}")
print(f"Traffic Level: {trafficlevel}")
print(f"estimated delivery time:{eta}")
```

```python
4 #cloud server load classification
def load_classify(server_loads_list):
  # Calculate the average load from the list of server loads
  if len(server_loads_list) == 0:
    print("Alert: No server load readings provided.")
    return "Unknown", 0.0

  average_load = sum(server_loads_list) / len(server_loads_list)

  # Classify server status based on the average load
  if average_load < 50:
    serverstatus = "Normal"
  elif 50 <= average_load <= 80:
    serverstatus = "Warning"
  else:
    serverstatus = "Critical"

  print(f"Server status: {serverstatus}")
  print(f"Average load: {average_load:.2f}")
  return serverstatus, average_load

# Sample CPU readings
cpu_readings = [45, 60, 70, 85, 90]

# Call the function with the sample data
classified_status, calculated_average = load_classify(cpu_readings)
```

```python
5 #smart classroom  resource usage monitor
def resource_monitor(projector,AC,lights):
  overused_resources = []
  if projector:
    overused_resources.append("Projector")
  if AC:
    overused_resources.append("AC")
  if lights:
    overused_resources.append("Lights")

  if len(overused_resources) > 1:
    energy_alert = "High energy consumption"
  elif len(overused_resources) == 1:
    energy_alert = "Moderate energy consumption"
  else:
    energy_alert = "Low energy consumption"

  print(f"overused resource: {', '.join(overused_resources) if overused_resources else 'None'}")
  print(f"energy alert: {energy_alert}")

# Sample input in dictionary format
resource_status = {
    "projector": False,
    "AC": True,
    "lights": False
}

# Call the function with dictionary values
resource_monitor(resource_status["projector"], resource_status["AC"], resource_status["lights"])
```

```python
6 #online event registration capacity
def event_controller(event_capacity,registration_list):
  confirmed_registrations = []
  waitlist_users = []

  for user in registration_list:
    if len(confirmed_registrations) < event_capacity:
      confirmed_registrations.append(user)
    else:
      waitlist_users.append(user)

  if len(confirmed_registrations) == event_capacity and len(waitlist_users) > 0:
    registration_status = "Waitlist only"
  elif len(confirmed_registrations) == event_capacity and len(waitlist_users) == 0:
    registration_status = "Closed (Full)"
  elif len(confirmed_registrations) < event_capacity:
    registration_status = "Open"
  else:
    registration_status = "Unknown"

  print(f"Confirmed Registrations: {', '.join(confirmed_registrations) if confirmed_registrations else 'None'}")
  print(f"Waitlist Users: {', '.join(waitlist_users) if waitlist_users else 'None'}")
  print(f"Registration Status: {registration_status}")

  return confirmed_registrations, waitlist_users, registration_status

# Sample Input
event_capacity_example = 5
registration_list_example = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace"]

# Call the function with sample input
confirmed, waitlist, status = event_controller(event_capacity_example, registration_list_example)
```
