```python
3 #find max and min numbers
numbers=[45,22,89,10,66]
max=numbers[0]
min=numbers[0]
for num in numbers:
    if num>max:
        max=num
    elif num<min:
        min=num

print(f"Max number is {max}")
print(f"Min number is {min}")
```

```python
7 #count charachter frequency
text="pythonp"
char_freq={}
for char in text:
  if char in char_freq:
    char_freq[char]+=1
  else:
    char_freq[char]=1
print(char_freq)
```
