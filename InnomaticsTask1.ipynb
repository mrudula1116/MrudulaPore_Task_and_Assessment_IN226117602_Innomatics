```python
#2 Pass/Fail Analyzer
marks=[45,78,90,33,60]
pass_count=0 #for storing count
fail_count=0 #for storing count

for mark in marks:
  if mark>=50:
    print("student is pass")
    pass_count=pass_count+1 #go on counting
  else:
    print("student is fail")
    fail_count=fail_count+1
print("total pass count is",pass_count)
print("total fail count is",fail_count)
```

```python
#simple data cleaner
names=["Alice","bob","CHARLIE"]
cleaned_names=[]
for name in names:
  cleaned_name=name.strip().lower() #removing leading trailing spaces and converting into lowercase
  cleaned_names.append(cleaned_name)
print(cleaned_names)
```

```python
#4 Message length analyzer
messages=["Hi","Welcome to the platform","OK"]
for message in messages:
  message_length=len(message) #calculates length
print(message_length)
if message_length>10:
  print("message is too long : flagged")
```

```python
#Error message detector
logs=["INFO","ERROR","WARNING","ERROR"]
error_count=0
for entry in logs:
  if entry=="ERROR": #find error named entry
    error_count=error_count+1
print("total error count is",error_count)
```
