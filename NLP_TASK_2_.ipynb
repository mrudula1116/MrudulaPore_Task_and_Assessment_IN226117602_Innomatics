```python
#text preprocessing
import pandas as pd
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Added to resolve LookupError for 'punkt_tab'
```

```python
#lowercase
def to_lowercase(text):
    return text.lower()
#punctuation removal
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))
#stopewords
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]
#tokeniztion
def tokenize_text(text):
    return word_tokenize(text)
#stemming , lemmtization
def apply_stemming(tokens):
    return [stemmer.stem(word) for word in tokens]
def apply_lemmatization(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]
#removing special charavchters
def remove_urls_special(text):
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # remove URLs
    text = re.sub(r'\n', ' ', text)  # remove newline chars
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # keep only alphabets
    return text
```

Now that the individual preprocessing functions are defined, we need to initialize the NLTK objects they rely on (lemmatizer, stemmer, and stopwords set). After that, we'll create a unified `preprocess_text` function to apply all these steps in sequence to the review data.

```python
# Initialize NLTK objects required by the preprocessing functions
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

print("NLTK Lemmatizer, Stemmer, and Stopwords initialized.")

# Define a comprehensive preprocessing pipeline function
def preprocess_text_pipeline(text):
    # Apply functions from cell eoXySCwUZT1q
    text = to_lowercase(text)
    text = remove_urls_special(text) # Should be applied before punctuation and tokenization
    text = remove_punctuation(text)
    tokens = tokenize_text(text)
    tokens = remove_stopwords(tokens)
    tokens = apply_lemmatization(tokens) # Using lemmatization as it's generally preferred over stemming for retaining meaning
    # If you also want to apply stemming, you can add: tokens = apply_stemming(tokens)
    return ' '.join(tokens)

print("Preprocessing pipeline function 'preprocess_text_pipeline' defined.")
```

Now, let's apply the `preprocess_text_pipeline` function to the 'review' column of your DataFrame and display the original and processed reviews to see the effects of the cleaning.

```python
# Load dataset - REMOVED: df was already loaded from data_file in an earlier cell (ur3s39owZH08).
# If you need to re-load, use: df = pd.read_csv(data_file)

# Apply preprocessing
# The correct preprocessing pipeline function is 'preprocess_text_pipeline'
# It does not take 'use_stemming' as an argument.
print("Applying preprocessing pipeline to the 'review' column... This may take a few moments.")
df['clean_review'] = df['review'].apply(lambda x: preprocess_text_pipeline(x))

# View result
print("\n--- Original vs. Processed Review Samples ---")
for i in range(5):
    print(f"\nOriginal Review {i+1}:\n{df['review'].iloc[i]}")
    print(f"Processed Review {i+1}:\n{df['clean_review'].iloc[i]}")

print("\nPreprocessing complete. Displaying the first few rows with the new 'clean_review' column:")
display(df[['review', 'clean_review', 'sentiment']].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer
# You can adjust parameters like max_features, min_df, max_df, ngram_range
count_vectorizer = CountVectorizer(max_features=5000) # Limiting to 5000 features for demonstration

# Fit and transform the 'clean_review' column
bow_matrix = count_vectorizer.fit_transform(df['clean_review'])

print("--- Bag of Words (BoW) Matrix Shape ---")
print(f"Shape of BoW matrix: {bow_matrix.shape}")

print("\n--- Sample BoW Feature Names (first 20) ---")
print(count_vectorizer.get_feature_names_out()[:20])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
# Similar to CountVectorizer, you can adjust parameters like max_features, min_df, max_df, ngram_range
tfidf_vectorizer = TfidfVectorizer(max_features=5000) # Limiting to 5000 features to match BoW

# Fit and transform the 'clean_review' column
tfidf_matrix = tfidf_vectorizer.fit_transform(df['clean_review'])

print("--- TF-IDF Matrix Shape ---")
print(f"Shape of TF-IDF matrix: {tfidf_matrix.shape}")

print("\n--- Sample TF-IDF Feature Names (first 20) ---")
print(tfidf_vectorizer.get_feature_names_out()[:20])
```

### Data Preparation for Model Training

Before training the models, we need to:
1.  **Encode the target variable**: Convert 'positive'/'negative' sentiment labels into numerical values.
2.  **Split the dataset**: Divide the data into training and testing sets to evaluate model performance on unseen data.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Encode the target variable 'sentiment'
label_encoder = LabelEncoder()
df['sentiment_encoded'] = label_encoder.fit_transform(df['sentiment'])

# Display the mapping
print("Sentiment Label Mapping:")
for i, label in enumerate(label_encoder.classes_):
    print(f"{label}: {i}")

Y = df['sentiment_encoded']

# Split data into training and testing sets
# Using BoW matrix as X for initial split
X_train_bow, X_test_bow, Y_train, Y_test = train_test_split(bow_matrix, Y, test_size=0.2, random_state=42, stratify=Y)

# Also split TF-IDF matrix using the same indices for consistency
X_train_tfidf, X_test_tfidf, _, _ = train_test_split(tfidf_matrix, Y, test_size=0.2, random_state=42, stratify=Y)

print(f"\nTraining set shape (BoW): {X_train_bow.shape}")
print(f"Testing set shape (BoW): {X_test_bow.shape}")
print(f"Training set shape (TF-IDF): {X_train_tfidf.shape}")
print(f"Testing set shape (TF-IDF): {X_test_tfidf.shape}")
print(f"Training target shape: {Y_train.shape}")
print(f"Testing target shape: {Y_test.shape}")
```

### Model Training and Evaluation

We will now train and evaluate the specified models (Logistic Regression, Naive Bayes, Decision Tree) using both Bag of Words (BoW) and TF-IDF features.

#### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

print("\n--- Logistic Regression with BoW Features ---")
logreg_bow = LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1) # Increased max_iter for convergence
logreg_bow.fit(X_train_bow, Y_train)
Y_pred_logreg_bow = logreg_bow.predict(X_test_bow)
print(f"Accuracy (BoW): {accuracy_score(Y_test, Y_pred_logreg_bow):.4f}")
print("Classification Report (BoW):")
print(classification_report(Y_test, Y_pred_logreg_bow, target_names=label_encoder.classes_))

print("\n--- Logistic Regression with TF-IDF Features ---")
logreg_tfidf = LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1)
logreg_tfidf.fit(X_train_tfidf, Y_train)
Y_pred_logreg_tfidf = logreg_tfidf.predict(X_test_tfidf)
print(f"Accuracy (TF-IDF): {accuracy_score(Y_test, Y_pred_logreg_tfidf):.4f}")
print("Classification Report (TF-IDF):")
print(classification_report(Y_test, Y_pred_logreg_tfidf, target_names=label_encoder.classes_))
```

#### 2. Naive Bayes (MultinomialNB)

```python
from sklearn.naive_bayes import MultinomialNB

print("\n--- Multinomial Naive Bayes with BoW Features ---")
mnb_bow = MultinomialNB()
mnb_bow.fit(X_train_bow, Y_train)
Y_pred_mnb_bow = mnb_bow.predict(X_test_bow)
print(f"Accuracy (BoW): {accuracy_score(Y_test, Y_pred_mnb_bow):.4f}")
print("Classification Report (BoW):")
print(classification_report(Y_test, Y_pred_mnb_bow, target_names=label_encoder.classes_))

print("\n--- Multinomial Naive Bayes with TF-IDF Features ---")
mnb_tfidf = MultinomialNB()
mnb_tfidf.fit(X_train_tfidf, Y_train)
Y_pred_mnb_tfidf = mnb_tfidf.predict(X_test_tfidf)
print(f"Accuracy (TF-IDF): {accuracy_score(Y_test, Y_pred_mnb_tfidf):.4f}")
print("Classification Report (TF-IDF):")
print(classification_report(Y_test, Y_pred_mnb_tfidf, target_names=label_encoder.classes_))
```

#### 3. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

print("\n--- Decision Tree with BoW Features ---")
dt_bow = DecisionTreeClassifier(random_state=42)
dt_bow.fit(X_train_bow, Y_train)
Y_pred_dt_bow = dt_bow.predict(X_test_bow)
print(f"Accuracy (BoW): {accuracy_score(Y_test, Y_pred_dt_bow):.4f}")
print("Classification Report (BoW):")
print(classification_report(Y_test, Y_pred_dt_bow, target_names=label_encoder.classes_))

print("\n--- Decision Tree with TF-IDF Features ---")
dt_tfidf = DecisionTreeClassifier(random_state=42)
dt_tfidf.fit(X_train_tfidf, Y_train)
Y_pred_dt_tfidf = dt_tfidf.predict(X_test_tfidf)
print(f"Accuracy (TF-IDF): {accuracy_score(Y_test, Y_pred_dt_tfidf):.4f}")
print("Classification Report (TF-IDF):")
print(classification_report(Y_test, Y_pred_dt_tfidf, target_names=label_encoder.classes_))
```
