```python
3 #Agriculture – Rainfall Adequacy Checker
def rainfall_checker(data,required_level):
  total_rainfall=0
  for amount in data:
    total_rainfall+=amount
  average_rainfall=total_rainfall/len(data)

  # Determine rainfall status and store it in a variable
  if average_rainfall>=required_level:
    rainfall_status="adequate rainfall"
  else:
    rainfall_status="inadequate rainfall"

  # Now return both the average rainfall and the status
  return average_rainfall,rainfall_status

rainfall_data=[10,20,30,40,50]
required_level=30

# Unpack the two returned values into separate variables
average_rainfall,rainfall_status_result=rainfall_checker(rainfall_data,required_level)

print(f"Average rainfall: {average_rainfall}")
print(f"Rainfall status: {rainfall_status_result}")
```

```python
10 #Smart Electricity Bill Analyzer

def calculate_bill(units):
    total_bill = 0
    if units <= 100:
        total_bill = units * 3
        print("price is rupee 3 ")
    elif units <= 200:
        total_bill = units * 5
        print("price is rupee 5")
    else:
        total_bill = units * 7
        print("price is rupee 7")
    return total_bill

def classify_usage(total_bill):
    if total_bill < 500:
        return "Low Usage"
    elif total_bill <= 1500:
        return "Moderate Usage"
    else:
        return "High Usage"

# Get user input for units consumed
units_str = int(input("Enter total units consumed: "))


# Calculate bill and classify usage
bill = calculate_bill(units_str)
usage_category = classify_usage(bill)

print(f"Total Bill: ₹{bill}")
print(f"Usage Classification: {usage_category}")
```
