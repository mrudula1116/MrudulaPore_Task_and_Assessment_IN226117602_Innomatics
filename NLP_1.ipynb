```python
import re
import string

def preprocess_text(text):
  text = text.lower() # Convert text to lowercase

  # Remove URLs and email-like patterns
  text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text)
  text = re.sub(r'\S*@\S*\s?', '', text)

  # Remove numbers
  text = re.sub(r'\d+', '', text)

  # Remove punctuation
  # Example: "Hello!!!" -> "Hello"
  text = text.translate(str.maketrans('', '', string.punctuation))


  text = re.sub(r'(.)\1{2,}', r'\1\1', text)

  # Remove extra spaces
  # Replaces multiple spaces with a single space and removes leading/trailing spaces
  text = re.sub(r'\s+', ' ', text).strip()

  # Remove very short tokens (words with length <= 2, keep "no", "not")
  tokens = text.split()
  meaningful_short_words = ['no', 'not']
  filtered_tokens = [word for word in tokens if len(word) > 2 or word in meaningful_short_words]
  text = ' '.join(filtered_tokens)

  return text
```

```python
example_text = "WOW!!! This is SOOOO GREAT!!! Visit http://example.com and email me at test@example.com for more info. I have 123 dogs and 45 cats. This is a very short text."
preprocessed_text = preprocess_text(example_text)
print(f"Original text: {example_text}")
print(f"Preprocessed text: {preprocessed_text}")
```

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in sample_inputs:
    preprocessed_text = preprocess_text(text)
    cleaned_tokens = preprocessed_text.split()
    print(f"Original Text: {text}")
    print(f"Cleaned Tokens: {cleaned_tokens}")
    print(f"Cleaned Sentence: {preprocessed_text}\n")
```
