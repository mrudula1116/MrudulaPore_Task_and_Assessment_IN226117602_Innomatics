```python
1 #smart light controller
class light_controller:
  def __init__(self,lightName,action):
    self.lightName=lightName
    self.action=action
  def light_control(self):
    print(f"{self.lightName} is {self.action}")

# Sample usage
bedroom_light = light_controller("Bedroom Light", "ON")
bedroom_light.light_control()
```

```python
#employee id card system
class student:
  def __init__(self,name,id,department):
    self.name=name
    self.id=id
    self.department=department
  def student_info(self):
    print("Employee ID info is")
    print(f"Name:{self.name}\nID:{self.id}\nDEPT:{self.department}")

# Sample usage
s = student("Rahul", "EMP102","IT")
s.student_info()
```

```python
#mobile contact record
class mobile:
  def __init__(self,name,number):
    self.name=name
    self.number=number
  def contact_info(self):
    print("Contact info is")
    print(f"Name:{self.name}\nNumber:{self.number}")


# Sample usage
s = mobile("sam", "9876543210")
s.contact_info()
```

```python
#product price tag generator
class product:
  def __init__(self,name,price):
    self.name=name
    self.price=price
  def product_info(self):
    print("product details:")
    print(f"Name:{self.name}\nPrice:{self.price}")
p=product("headphones",2499)
p.product_info()
```

```python
#movie rating display system
class movie:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating
  def movie_info(self):
    print("Movie info:")
    print(f"Name:{self.name}\nRating:{self.rating}/5")
m=movie("Inception",4.8)
m.movie_info()
```

```python
#delivery address manager
class delivery_manager:
  def __init__(self,name,address):
    self.name=name
    self.address=address
  def delivery_info(self):
    print("Delivery details:")
    print(f"Name:{self.name}\nAddress:{self.address}")
d=delivery_manager("suman","hyderabad")
d.delivery_info()
```
